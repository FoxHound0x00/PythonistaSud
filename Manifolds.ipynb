```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def create_manifold_points(num_points=1000):
    theta = np.linspace(0, 2 * np.pi, num_points)
    x = 2 * np.cos(theta)
    y = np.sin(theta)
    z = 0.1 * theta
    return x, y, z

def plot_manifold(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y, z, label='Manifold')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('2D Manifold in 3D Space')
    ax.legend()
    equation_text = r'$x = 2 \cos(\theta)$' + '\n' + r'$y = \sin(\theta)$' + '\n' + r'$z = 0.1 \theta$'
    ax.text2D(0.05, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')

    plt.show()

x_manifold, y_manifold, z_manifold = create_manifold_points()
plot_manifold(x_manifold, y_manifold, z_manifold)
```

```python
def create_torus_points(num_points=1000, R=3, r=1):
    theta = np.linspace(0, 2 * np.pi, num_points)
    phi = np.linspace(0, 2 * np.pi, num_points)
    theta, phi = np.meshgrid(theta, phi)
    x = (R + r * np.cos(phi)) * np.cos(theta)
    y = (R + r * np.cos(phi)) * np.sin(theta)
    z = r * np.sin(phi)

    return x, y, z

def plot_torus(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('3D Torus Manifold')

    equation_text = r'$x = (R + r \cos(\phi)) \cos(\theta)$' + '\n' + r'$y = (R + r \cos(\phi)) \sin(\theta)$' + '\n' + r'$z = r \sin(\phi)$'
    ax.text2D(1,1, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')
    plt.show()

x_torus, y_torus, z_torus = create_torus_points()
plot_torus(x_torus, y_torus, z_torus)
```

```python
def klein_bottle(u, v):
    x = -2 + 2 * np.cos(u) - np.sin(u/2) * np.cos(v)
    y = -2 * np.sin(u) - np.sin(u/2) * np.sin(v)
    z = np.cos(u/2) * np.sin(v)
    return x, y, z

def create_klein_bottle_points(num_points=500):
    u = np.linspace(0, 2 * np.pi, num_points)
    v = np.linspace(0, 2 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = klein_bottle(u, v)
    return x, y, z

def plot_klein_bottle(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Klein Bottle')

    equation_text = r'$x = -2 + 2 \cos(u) - \frac{\sin(u)}{2} \cos(v)$' + '\n' + \
                    r'$y = -2 \sin(u) - \frac{\sin(u)}{2} \sin(v)$' + '\n' + \
                    r'$z = \frac{\cos(u)}{2} \sin(v)$'
    ax.text2D(1,1, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')
    plt.show()

x_klein, y_klein, z_klein = create_klein_bottle_points()
plot_klein_bottle(x_klein, y_klein, z_klein)
```

```python
def twisted_torus(u, v, R=3, r=1, twist_factor=2):
    x = (R + r * np.cos(twist_factor * u)) * np.cos(v)
    y = (R + r * np.cos(twist_factor * u)) * np.sin(v)
    z = r * np.sin(twist_factor * u)
    return x, y, z

def create_twisted_torus_points(num_points=500):
    u = np.linspace(0, 2 * np.pi, num_points)
    v = np.linspace(0, 2 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = twisted_torus(u, v)
    return x, y, z

def plot_twisted_torus(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Twisted Torus')
    equation_text = (r"$x = (R + r \cos(ku)) \cos(v)$" + '\n' +
                     r"$y = (R + r \cos(ku)) \sin(v)$" + '\n' +
                     r"$z = r \sin(ku)$")
    ax.text2D(1,1, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')


    plt.show()

x_twisted_torus, y_twisted_torus, z_twisted_torus = create_twisted_torus_points()
plot_twisted_torus(x_twisted_torus, y_twisted_torus, z_twisted_torus)
```

```python
def mobius_strip(u, v, width=1, twist_factor=2):
    x = (width + v * np.cos(u / 2)) * np.cos(u)
    y = (width + v * np.cos(u / 2)) * np.sin(u)
    z = v * np.sin(u / 2)
    return x, y, z

def create_mobius_strip_points(num_points=500):
    u = np.linspace(0, 2 * np.pi, num_points)
    v = np.linspace(-1, 1, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = mobius_strip(u, v)
    return x, y, z

def plot_mobius_strip(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Mbius Strip')
    equation_text = (r"$x = (w + v \cos(u/2)) \cos(u)$" + '\n' +
                     r"$y = (w + v \cos(u/2)) \sin(u)$" + '\n' +
                     r"$z = v \sin(u/2)$")
    ax.text2D(1,1, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')


    plt.show()

x_mobius, y_mobius, z_mobius = create_mobius_strip_points()
plot_mobius_strip(x_mobius, y_mobius, z_mobius)
```

```python
def trefoil_knot(u, v, scale=1):
    x = scale * (np.sin(u) + 2 * np.sin(2 * u))
    y = scale * (np.cos(u) - 2 * np.cos(2 * u))
    z = scale * (-np.sin(3 * u))
    return x, y, z

def create_trefoil_knot_points(num_points=500):
    u = np.linspace(0, 4 * np.pi, num_points)
    v = np.linspace(0, 2 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = trefoil_knot(u, v)
    return x, y, z

def plot_trefoil_knot(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Trefoil Knot')
    equation_text = (r"$x = \sin(u) + 2 \sin(2u)$" + '\n' +
                     r"$y = \cos(u) - 2 \cos(2u)$" + '\n' +
                     r"$z = -\sin(3u)$")
    ax.text2D(1,1, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')


    plt.show()

x_trefoil, y_trefoil, z_trefoil = create_trefoil_knot_points()
plot_trefoil_knot(x_trefoil, y_trefoil, z_trefoil)
```

```python
def hyperbolic_paraboloid(u, v, a=1, b=1):
    x = a * u
    y = b * v
    z = u**2 - v**2
    return x, y, z

def create_hyperbolic_paraboloid_points(num_points=500):
    u = np.linspace(-1, 1, num_points)
    v = np.linspace(-1, 1, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = hyperbolic_paraboloid(u, v)
    return x, y, z

def plot_hyperbolic_paraboloid(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Hyperbolic Paraboloid')

    equation_text = (r"$x = a \cdot u$" + '\n' +
                     r"$y = b \cdot v$" + '\n' +
                     r"$z = u^2 - v^2$")
    ax.text2D(1,1, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')


    plt.show()

x_hyperbolic, y_hyperbolic, z_hyperbolic = create_hyperbolic_paraboloid_points()
plot_hyperbolic_paraboloid(x_hyperbolic, y_hyperbolic, z_hyperbolic)
```

```python
def torus_knot(u, v, p=3, q=2, R=1, r=0.3):
    x = (R + r * np.cos(q * u)) * np.cos(p * u)
    y = (R + r * np.cos(q * u)) * np.sin(p * u)
    z = r * np.sin(q * u)
    return x, y, z

def create_torus_knot_points(num_points=500):
    u = np.linspace(0, 2 * np.pi, num_points)
    v = np.linspace(0, 2 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = torus_knot(u, v)
    return x, y, z

def plot_torus_knot(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Torus Knot')
    equation_text = (r"$x = (R + r \cdot \cos(q \cdot u)) \cdot \cos(p \cdot u)$" + '\n' +
                     r"$y = (R + r \cdot \cos(q \cdot u)) \cdot \sin(p \cdot u)$" + '\n' +
                     r"$z = r \cdot \sin(q \cdot u)$")
    ax.text2D(0.05, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')

    plt.show()

x_torus_knot, y_torus_knot, z_torus_knot = create_torus_knot_points()
plot_torus_knot(x_torus_knot, y_torus_knot, z_torus_knot)
```

```python
def twisted_cubic(t, a=1):
    x = a * t
    y = a * t**2
    z = a * t**3
    return x, y, z

def create_twisted_cubic_points(num_points=500):
    t = np.linspace(-1, 1, num_points)
    x, y, z = twisted_cubic(t)
    return x, y, z

def plot_twisted_cubic(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y, z, label='Twisted Cubic')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Twisted Cubic Curve')
    ax.legend()
    plt.show()

x_twisted_cubic, y_twisted_cubic, z_twisted_cubic = create_twisted_cubic_points()
plot_twisted_cubic(x_twisted_cubic, y_twisted_cubic, z_twisted_cubic)
```

```python
def spherical_helix(t, R=1, a=0.1):
    x = R * np.cos(a * t) * np.cos(t)
    y = R * np.cos(a * t) * np.sin(t)
    z = R * np.sin(a * t)
    return x, y, z

def create_spherical_helix_points(num_points=500):
    t = np.linspace(0, 10 * np.pi, num_points)
    x, y, z = spherical_helix(t)
    return x, y, z

def plot_spherical_helix(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y, z, label='Spherical Helix')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Spherical Helix')
    ax.legend()
    plt.show()

x_spherical_helix, y_spherical_helix, z_spherical_helix = create_spherical_helix_points()
plot_spherical_helix(x_spherical_helix, y_spherical_helix, z_spherical_helix)
```

```python
def helicoid(u, v, a=1, b=1):
    x = a * u * np.cos(v)
    y = a * u * np.sin(v)
    z = b * v
    return x, y, z

def create_helicoid_points(num_points=500):
    u = np.linspace(-2 * np.pi, 2 * np.pi, num_points)
    v = np.linspace(0, 4 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = helicoid(u, v)
    return x, y, z

def plot_helicoid(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Helicoid')
    equation_text = (r"$x = a \cdot u \cdot \cos(v)$" + '\n' +
                     r"$y = a \cdot u \cdot \sin(v)$" + '\n' +
                     r"$z = b \cdot v$")
    ax.text2D(0.05, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')


    plt.show()

x_helicoid, y_helicoid, z_helicoid = create_helicoid_points()
plot_helicoid(x_helicoid, y_helicoid, z_helicoid)
```

```python
def hyperboloid_one_sheet(u, v, a=1, b=1, c=1):
    x = a * np.sinh(u) * np.cos(v)
    y = b * np.sinh(u) * np.sin(v)
    z = c * np.cosh(u)
    return x, y, z

def create_hyperboloid_one_sheet_points(num_points=500):
    u = np.linspace(-2, 2, num_points)
    v = np.linspace(0, 2 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = hyperboloid_one_sheet(u, v)
    return x, y, z

def plot_hyperboloid_one_sheet(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Hyperboloid of One Sheet')
    equation_text = (r"$x = a \sinh(u) \cos(v)$" + '\n' +
                     r"$y = b \sinh(u) \sin(v)$" + '\n' +
                     r"$z = c \cosh(u)$")
    ax.text2D(1,1, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')



    plt.show()

x_hyperboloid_one_sheet, y_hyperboloid_one_sheet, z_hyperboloid_one_sheet = create_hyperboloid_one_sheet_points()
plot_hyperboloid_one_sheet(x_hyperboloid_one_sheet, y_hyperboloid_one_sheet, z_hyperboloid_one_sheet)
```

```python
def mobius_band(u, v, width=1, twist_factor=2):
    x = (width + v * np.cos(u / 2)) * np.cos(u)
    y = (width + v * np.cos(u / 2)) * np.sin(u)
    z = v * np.sin(u / 2)
    return x, y, z

def create_mobius_band_points(num_points=500):
    u = np.linspace(0, 2 * np.pi, num_points)
    v = np.linspace(-1, 1, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = mobius_band(u, v)
    return x, y, z

def plot_mobius_band(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Mbius Band')
    equation_text = (r"$x = (w + v \cos(u/2)) \cos(u)$" + '\n' +
                     r"$y = (w + v \cos(u/2)) \sin(u)$" + '\n' +
                     r"$z = v \sin(u/2)$")
    ax.text2D(1, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')


    plt.show()

x_mobius_band, y_mobius_band, z_mobius_band = create_mobius_band_points()
plot_mobius_band(x_mobius_band, y_mobius_band, z_mobius_band)
```

```python
def enneper_surface(u, v):
    x = u - (u**3) / 3 + u * (v**2 - (u**2) / 3)
    y = v - (v**3) / 3 + v * (u**2 - (v**2) / 3)
    z = u**2 - v**2
    return x, y, z

def create_enneper_surface_points(num_points=500):
    u = np.linspace(-2, 2, num_points)
    v = np.linspace(-2, 2, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = enneper_surface(u, v)
    return x, y, z

def plot_enneper_surface(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Enneper Surface')

    equation_text = (r"$x = u$" + '\n' +
                     r"$y = v$" + '\n' +
                     r"$z = u^2 v$")
    ax.text2D(0.05, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')


    plt.show()

x_enneper, y_enneper, z_enneper = create_enneper_surface_points()
plot_enneper_surface(x_enneper, y_enneper, z_enneper)
```

```python
def bohemian_dome(u, v):
    x = u * np.cos(v)
    y = u * np.sin(v)
    z = 0.2 * u**2 + np.sin(5 * v)
    return x, y, z

def create_bohemian_dome_points(num_points=1000):
    u = np.linspace(0, 2, num_points)
    v = np.linspace(0, 2 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = bohemian_dome(u, v)
    return x, y, z

def plot_bohemian_dome(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Bohemian Dome')

    equation_text = (r"$x = u$" + '\n' +
                     r"$y = v$" + '\n' +
                     r"$z = u^2 v$")
    ax.text2D(0.05, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')

    plt.show()

x_bohemian_dome, y_bohemian_dome, z_bohemian_dome = create_bohemian_dome_points()
plot_bohemian_dome(x_bohemian_dome, y_bohemian_dome, z_bohemian_dome)
```

```python
def ellipsoid(u, v, a=1, b=2, c=3):
    x = a * np.sin(u) * np.cos(v)
    y = b * np.sin(u) * np.sin(v)
    z = c * np.cos(u)
    return x, y, z

def create_ellipsoid_points(num_points=500):
    u = np.linspace(0, np.pi, num_points)
    v = np.linspace(0, 2 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = ellipsoid(u, v)
    return x, y, z

def plot_ellipsoid(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Ellipsoid')

    equation_text = (r"$x = A \sin(u) \cos(v)$" + '\n' +
                     r"$y = B \sin(u) \sin(v)$" + '\n' +
                     r"$z = C \cos(u)$")
    ax.text2D(0.05, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')

    plt.show()

x_ellipsoid, y_ellipsoid, z_ellipsoid = create_ellipsoid_points()
plot_ellipsoid(x_ellipsoid, y_ellipsoid, z_ellipsoid)
```

```python
def seashell_surface(u, v):
    a = 1
    b = 1.5
    x = a * (1 - v/(2*np.pi)) * np.cos(u) * (1 + np.cos(v))
    y = a * (1 - v/(2*np.pi)) * np.sin(u) * (1 + np.cos(v))
    z = b * v/(2*np.pi) + a * (1 - v/(2*np.pi)) * np.sin(v)
    return x, y, z

def create_seashell_surface_points(num_points=500):
    u = np.linspace(0, 2 * np.pi, num_points)
    v = np.linspace(0, 2 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = seashell_surface(u, v)
    return x, y, z

def plot_seashell_surface(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Seashell Surface')

    equation_text = (r"$x = a(1 - \frac{v}{2\pi})\cos(u)(1 + \cos(v))$" + '\n' +
                     r"$y = a(1 - \frac{v}{2\pi})\sin(u)(1 + \cos(v))$" + '\n' +
                     r"$z = b\frac{v}{2\pi} + a(1 - \frac{v}{2\pi})\sin(v)$")
    ax.text2D(1, 1, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')

    plt.show()

x_seashell, y_seashell, z_seashell = create_seashell_surface_points()
plot_seashell_surface(x_seashell, y_seashell, z_seashell)
```

```python
def clifford_torus(u, v, a=2, b=3, c=1, d=1):
    x = (a + np.cos(b * u)) * np.cos(c * v)
    y = (a + np.cos(b * u)) * np.sin(c * v)
    z = np.sin(b * u) * np.sin(d * v)
    return x, y, z

def create_clifford_torus_points(num_points=500):
    u = np.linspace(0, 2 * np.pi, num_points)
    v = np.linspace(0, 2 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = clifford_torus(u, v)
    return x, y, z

def plot_clifford_torus(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Clifford Torus')
    equation_text = (r"$x = (a + \cos(bu))\cos(cv)$" + '\n' +
                     r"$y = (a + \cos(bu))\sin(cv)$" + '\n' +
                     r"$z = \sin(bu)\sin(dv)$")
    ax.text2D(1,1, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')

    plt.show()

x_clifford_torus, y_clifford_torus, z_clifford_torus = create_clifford_torus_points()
plot_clifford_torus(x_clifford_torus, y_clifford_torus, z_clifford_torus)
```

```python
def roman_surface(u, v):
    x = np.sin(u) * np.cos(v)
    y = np.sin(u) * np.sin(v)
    z = np.cos(u) + np.log(np.tan(u/2))
    return x, y, z

def create_roman_surface_points(num_points=500):
    u = np.linspace(0.01, np.pi - 0.01, num_points)
    v = np.linspace(0, 2 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = roman_surface(u, v)
    return x, y, z

def plot_roman_surface(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Roman Surface')
    equation_text = (r"$x = \sin(u) \cos(v)$" + '\n' +
                     r"$y = \sin(u) \sin(v)$" + '\n' +
                     r"$z = \cos(u) + \log(\tan(u/2))$")
    ax.text2D(0.05, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')

    plt.show()

x_roman_surface, y_roman_surface, z_roman_surface = create_roman_surface_points()
plot_roman_surface(x_roman_surface, y_roman_surface, z_roman_surface)
```

```python
def saddle_tower(u, v):
    x = u * np.cos(v)
    y = u * np.sin(v)
    z = u**2 - v**2
    return x, y, z

def create_saddle_tower_points(num_points=500):
    u = np.linspace(-2, 2, num_points)
    v = np.linspace(-2 * np.pi, 2 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = saddle_tower(u, v)
    return x, y, z

def plot_saddle_tower(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Saddle Tower')

    equation_text = r"$x = u \cos(v)$" + '\n' + r"$y = u \sin(v)$" + '\n' + r"$z = u^2 - v^2$"
    ax.text2D(0.05, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')
    plt.show()

x_saddle_tower, y_saddle_tower, z_saddle_tower = create_saddle_tower_points()
plot_saddle_tower(x_saddle_tower, y_saddle_tower, z_saddle_tower)
```

```python
def swallowtail_catastrophe(u, v):
    x = u
    y = v
    z = u**3 - 3 * u * v**2
    return x, y, z

def create_swallowtail_catastrophe_points(num_points=500):
    u = np.linspace(-2, 2, num_points)
    v = np.linspace(-2, 2, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = swallowtail_catastrophe(u, v)
    return x, y, z

def plot_swallowtail_catastrophe(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Swallowtail Catastrophe')

    equation_text = r"$x = u$" + '\n' + r"$y = v$" + '\n' + r"$z = u^3 - 3uv^2$"
    ax.text2D(0.05, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')
    plt.show()

x_swallowtail, y_swallowtail, z_swallowtail = create_swallowtail_catastrophe_points()
plot_swallowtail_catastrophe(x_swallowtail, y_swallowtail, z_swallowtail)
```

```python
def riemann_surface(u, v):
    x = u
    y = v
    z = np.sin(u) * np.sin(v) + 0.2 * u
    return x, y, z

def create_riemann_surface_points(num_points=500):
    u = np.linspace(0, 2 * np.pi, num_points)
    v = np.linspace(0, 2 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = riemann_surface(u, v)
    return x, y, z

def plot_riemann_surface(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title("Riemann's Minimal Surface")

    equation_text = (r"$x = u$" + '\n' +
                     r"$y = v$" + '\n' +
                     r"$z = \sin(u) \sin(v) + 0.2u$")
    ax.text2D(0.05, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')

    plt.show()

x_riemann_surface, y_riemann_surface, z_riemann_surface = create_riemann_surface_points()
plot_riemann_surface(x_riemann_surface, y_riemann_surface, z_riemann_surface)
```

```python
def bryants_surface(u, v):
    x = np.cos(u) * (np.cos(v) + np.log(np.tan(v/2)))
    y = np.sin(u) * (np.cos(v) + np.log(np.tan(v/2)))
    z = u + np.sin(v)
    return x, y, z

def create_bryants_surface_points(num_points=500):
    u = np.linspace(0.01, np.pi - 0.01, num_points)
    v = np.linspace(0, 2 * np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = bryants_surface(u, v)
    return x, y, z

def plot_bryants_surface(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title("Bryant's Surface")

    equation_text = (r"$x = \cos(u) \left(\cos(v) + \log\left(\tan\left(\frac{v}{2}\right)\right)\right)$" + '\n' +
                     r"$y = \sin(u) \left(\cos(v) + \log\left(\tan\left(\frac{v}{2}\right)\right)\right)$" + '\n' +
                     r"$z = u + \sin(v)$")
    ax.text2D(0.05, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')
    plt.show()

x_bryants_surface, y_bryants_surface, z_bryants_surface = create_bryants_surface_points()
plot_bryants_surface(x_bryants_surface, y_bryants_surface, z_bryants_surface)
```

```python
def kuens_surface(u, v):
    a = 0.2
    b = 0.3
    x = a * (np.cos(u) * np.cos(v) + np.sqrt(2) * np.sin(u) * np.sin(2*v)) * np.cos(u) / (1 + np.sqrt(2) * np.sin(3*u) * np.sin(2*v))
    y = a * (np.cos(u) * np.cos(v) + np.sqrt(2) * np.sin(u) * np.sin(2*v)) * np.sin(u) / (1 + np.sqrt(2) * np.sin(3*u) * np.sin(2*v))
    z = b * np.cos(u) * (1 + np.sqrt(2) * np.sin(3*u) * np.sin(2*v)) / (1 + np.sqrt(2) * np.sin(u) * np.sin(2*v))
    return x, y, z

def create_kuens_surface_points(num_points=500):
    u = np.linspace(0, 2 * np.pi, num_points)
    v = np.linspace(0, np.pi, num_points)
    u, v = np.meshgrid(u, v)
    x, y, z = kuens_surface(u, v)
    return x, y, z

def plot_kuens_surface(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title("Kuen's Surface")
    equation_text = (r"$x = \frac{a(\cos(u)\cos(v) + \sqrt{2}\sin(u)\sin(2v))\cos(u)}{1 + \sqrt{2}\sin(3u)\sin(2v)}$" + '\n' +
                     r"$y = \frac{a(\cos(u)\cos(v) + \sqrt{2}\sin(u)\sin(2v))\sin(u)}{1 + \sqrt{2}\sin(3u)\sin(2v)}$" + '\n' +
                     r"$z = \frac{b\cos(u)(1 + \sqrt{2}\sin(3u)\sin(2v))}{1 + \sqrt{2}\sin(u)\sin(2v)}$")
    ax.text2D(1.2,1, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')



    plt.show()

x_kuens_surface, y_kuens_surface, z_kuens_surface = create_kuens_surface_points()
plot_kuens_surface(x_kuens_surface, y_kuens_surface, z_kuens_surface)
```
